```python
import pandas as pd
import numpy as np
import sys
import os
from pathlib import Path
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.impute import SimpleImputer, KNNImputer, IterativeImputer
```

```python
filepath = os.path.join('data/test.csv')
test_csv = pd.read_csv(filepath)
test_csv
```

```python
#one-hot encoding
pd.get_dummies(test_csv[['Name']],prefix='Name')
```

```python
#label encoding
encoding = LabelEncoder()
encoding.fit_transform(test_csv['Name'])
```

```python
# Standard scaling
scalar = StandardScaler()
num_col = test_csv.select_dtypes(include = ['number']).columns
data_scaled = scalar.fit_transform(test_csv[num_col])
data_scaled
```

```python
# KNN imputer
knn = KNNImputer()
knn.fit_transform(test_csv[['Age']])
```

```python
# Iterative Imputer
Iterative_imputer = IterativeImputer()
Iterative_imputer.fit_transform(test_csv[['Age']])
```

```python
# mean,median,mode
test_csv[['Age']].mean()
```
